```python
covid_china = covid_data[covid_data['Country/Region'] == 'China']
covid_china
```

```python
covid_china = covid_china.groupby('Province/State')[['Confirmed', 'Deaths', 'Recovered', 'Active']].sum().reset_index()
covid_china
```

```python
covid_china.sort_values('Confirmed', ascending=False)
```

```python
import numpy as np


df_covid = covid_data.copy()

df_covid['Location'] = np.where(
  covid_data['Province/State'].notna(),
  covid_data['Country/Region'] + '_' + covid_data['Province/State'],
  covid_data['Country/Region']
)

df_covid
```

```python
df_covid.apply('Location')
```

```python
df_covid.drop('Province/State', axis=1, inplace=True)
df_covid
```

```python
df_covid.columns
```

```python
df_world = pd.read_csv(f'{path}/world_data.csv', encoding='latin1')
df_world
```

```python
df_world_filtered = df_world[['Continent', 'Country', 'Population']]
df_world_filtered
```

```python
df_merged = pd.merge(
    df_covid,
    df_world_filtered,
    left_on='Country/Region',
    right_on='Country',
    how='inner'
)

df_world_covid = df_merged[['Country', 'Population', 'Continent', 'Deaths']]
df_world_covid['Percent-Deaths'] = (df_world_covid['Deaths'] / df_world_covid['Population']) * 100
```

```python
df_world_covid.sort_values('Percent-Deaths', ascending=False)
df_world_covid
```

```python
df_world_continent = df_world_covid.groupby('Continent').sum().reset_index()
df_world_continent.drop('Country', axis=1, inplace=True)
df_world_continent.sort_values('Percent-Deaths', ascending=False)
```
